```python
import pandas as pd
df_wine = pd.read_csv("./data/winemag-data-130k-v2.csv")
```

```python
#Visualización de la data
df_wine.head(5)
```

```python
# EXPLORACIÓN DE LA DATA
#Visualicemos las 10 primeras filas
df_wine.head(10)

#Visualicemos las 10 últimas filas
df_wine.tail(10)

#Veamos el tipo de dato de cada columna
df_wine.dtypes

#Veamos cuales son los nombres de las columnas (variables)
df_wine.columns
#Resumen de mis variables
df_wine.describe()

# Veamos los países únicos
df_wine['country'].unique()
```

```python
#Nombremos a la primera columna que no tiene nombre "Unnamed: 0"
df=df_wine.rename(columns={'Unnamed: 0':'nombre_relleno'})
df.head(1)
# más adelante eliminaremos esta columna
```

```python
#Eliminamos esta columna porque no la usaremos
df=df.drop(['nombre_relleno'],axis=1)
df
```

```python
#Cambiamos algunos nombres a nuestras columnas
df=df.rename(columns={'country':'pais','description':'descripcion','points':'puntos','price':'precio'})
df.head(1)
```

```python
# Si el siguiente año todas las economías del mundo sufren una presión inflacionaria de 25%
#Cual sería el precio el siguiente año
df['precio_2025']=df['precio']*1.25
df
```

```python
#Definimos una función para determinar si el precio de un vino es :
#Precio económico
#Precio moderado
#precio caro
#precio muy caro
def etiqueta_precio(precio:float):
    if precio<100:
        return 'Precio económico'
    elif precio<600 and precio>=100:
        return 'Precio moderado'
    elif precio<1500 and precio>=600:
        return 'Precio caro'
    elif precio>=1500:
        return 'Precio muy caro'
etiqueta_precio(1200)
```

```python
df['nivel_precio']=df['precio'].apply(etiqueta_precio)
df.head(2)
```

```python
#Consideramos el siguiente diccionario
diccionario={
'Pais' : ['Morocco','South Africa','Argentina','Brazil','Canada','Chile','Mexico','Peru','Uruguay','US','India','Lebanon','Armenia','Austria','Bosnia and Herzegovina','Bulgaria','Croatia','Cyprus','Czech Republic','England','France','Georgia','Germany','Greece','Hungary','Israel','Italy','Luxembourg','Macedonia','Moldova','Portugal','Romania','Serbia','Slovakia','Slovenia','Spain','Switzerland','Turkey','Ukraine','Australia','New Zealand'],
'Continente':['Africa','Africa','América','América','América','América','América','América','América','América','Asia','Asia','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Europa','Oceanía','Oceanía']
}
df_pais=pd.DataFrame(diccionario)
df_pais
```

```python
df_merge=pd.merge(df,df_pais,left_on='pais',right_on="Pais")
df_merge
```

```python
#REPORTE 1
reporte1=df_merge.groupby('Continente')['puntos'].agg(['max'])
reporte1.sort_values(by=['max'],ascending=[False])
```

```python
#REPORTE 2
reporte2=df_merge.groupby('Pais').agg({'precio':'mean','puntos': 'sum'})
reporte2.sort_values(by=['puntos'],ascending=[False])
```

```python
#REPORTE 3
#Reporte de la cantidad de vendedores de vino por país y continente
reporte3=df_merge.groupby(['Continente','Pais']).agg({'pais': 'count'})
reporte3.sort_values(by=['Continente','pais'],ascending=[True,False])
```

```python
#REPORTE 4
#Este reporte muestras el precio promedio y la cantidad de vendedores de vino por nivel de precio
reporte4=df_merge.groupby('nivel_precio').agg({'precio':'mean','pais': 'count'})
reporte4.sort_values(by=['precio'],ascending=[False])
```

```python
#Enviaremos a gerencia nuestro reporte 4
ruta="./data/ReporteparaGerencia.xlsx"
reporte4.to_excel(ruta, index=False)
```
